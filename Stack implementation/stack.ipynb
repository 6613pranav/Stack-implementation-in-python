# Stack 

 * Linear data structure
 * stores data in LIFO /FILO
 * Elements are added and removed at end only
 * All the operation are done in space complexiety O(1)
 

## Implementation of stack in python
  #### It can be implemented with following methods
   * List
   * collection.deque
   * queue.LifoQueue

## 1. Implementation by List

  * List has two method 
   * pop() :- Deletes the last element from list
   * append() :- inserts a new item at end
   
  ### Problem with list 
   * when the size of list grows  bigger than the memory allocated by python 
   * the OS need to allocate new memory for append() method which takes extra time
   * Time complexiety  :  O(n) 
   

```python
# code
stack=[]
j=1
def push(e):
    stack.append(e)
    print('\n')
    

def pop():
    e=stack.pop()
    print('Popped element is {} : '.format(e))
    print('\n')
    

def stack_status():
    li2=stack
    li2.reverse()
    for i in li2:
        print('Elements of stack is :',i)
    print('\n')

while j!=0:
    e=input('Enter your choice push / pop / display_stack :- ')
    print('\n')
    if e== 'push':
        a=input('Enter the element ')
        if a.isdigit():
            push(int(a))
            
            
        else:
            push(a)
               
    elif e == 'pop':
        pop()
    
    else:
        stack_status()
    
    j=int(input('Enter 0 to terminate else press anything ..\n'))
        
```

## Implementation by collection.deque
  * Preferred over list because its speed of operation is fast
  * Time Complexiety :  O(1)   { to append or pop an element }
 
  ### We can use following methods
      suppose we have stack named stack_deque
      1. stack_deque.pop() 
       * Popping last element 
    
      2. len(stack_deque)
         * Finding the length of stack_deque
       
      3. stack_deque.append()
        * Inserting an element at last
    
      4. stack_deque.reverse()
       * Reversing the stack_deque
     
    
  

```python
# code

from collections import deque 

stack2 = deque() 

#---------------------appending element-------#
stack2.append('a') 
stack2.append('b') 
stack2.append('c') 

print('Initial stack:') 
print(stack2) 

# --------------Displaying Stack---------------#
print('\n\n Displaying stack ')
rev2=stack2
rev2.reverse()
for i in rev2:
    print(i)


#--------------popping elements from stack-----#
print('\n\nElements popped from stack:') 
print(stack2.pop()) 
print(stack2.pop()) 
print(stack2.pop()) 

print('\nStack after elements are popped:') 
print(stack2) 

# uncommenting print(stack.pop()) 
# will cause an IndexError 
# as the stack is now empty 
```

# Implementation  using queue module

## Methods used in queue module to implement stack
  
  * maxsize – Number of items allowed in the queue.
  
  
  * empty() – Return True if the queue is empty, False otherwise.
  
  
  * full() – Return True if there are maxsize items in the queue. 
     * If the queue was initialized with maxsize=0 (the default),then full() never returns True.
  
  
  * get() – Remove and return an item from the queue. If queue is empty, wait until an item is available.
  
  
  * get_nowait() – Return an item if one is immediately available, else raise QueueEmpty.
  
  
  * put(item) – Put an item into the queue. If the queue is full, wait until a free slot is available before adding the item.
  
  
  * put_nowait(item) – Put an item into the queue without blocking.
  
  
  * qsize() – Return the number of items in the queue. If no free slot is immediately available, raise QueueFull.

```python
from queue import LifoQueue 

# Initializing a stack 
stack3 = LifoQueue(maxsize = 3) 

# qsize() show the number of elements 
# in the stack 
print(stack3.qsize()) 

# put() function to push 
# element in the stack 
stack3.put('a') 
stack3.put('b') 
stack3.put('c') 

print("Full: ", stack3.full()) 
print("Size: ", stack3.qsize()) 

# get() fucntion to pop 
# element from stack in 
# LIFO order 
print('\nElements poped from the stack') 
print(stack3.get()) 
print(stack3.get()) 
print(stack3.get()) 

print("\nEmpty: ", stack3.empty()) 
```
